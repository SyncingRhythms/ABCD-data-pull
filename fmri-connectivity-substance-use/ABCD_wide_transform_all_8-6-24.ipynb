```python
def assign_labels(df, column_name):
    """
    Check if values in a column are numeric and assign labels accordingly.
    
    Parameters:
    - df: The DataFrame containing the column.
    - column_name: The name of the column to check.
    
    Returns:
    - A DataFrame with the updated column containing the assigned labels.
    """
    try:
        # Attempt to convert the value to numeric
        pd.to_numeric(df[column_name], errors='raise')
        label = 'float'
    except (ValueError, TypeError):
        label = 'object'

    # Apply the labeling function to the specified column
    df[column_name] = df[column_name].astype(label)
    return df
```

```python
for col in rs_wide.columns:
    rs_wide = assign_labels(rs_wide, col)
```
